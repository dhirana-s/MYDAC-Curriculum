```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Desktop/Netflix_Userbase.csv')
```

```python
df.head()
```

```python
#Cleaning data: 1) remove nulls, 2) deal with outliers, 3) remove duplicates
df.isnull().sum()
#verified that there are no null rows
```

```python
df['Monthly Revenue'].value_counts().sort_index().plot(kind='line', title='Sales per Month')
plt.rcParams['figure.figsize'] = (14, 6)
plt.show()
```

```python
df['Subscription Type'].value_counts().plot(kind='bar', title='Most Popular Subscription Type')
plt.rcParams['figure.figsize'] = (14, 6)
plt.xticks(rotation=45)
plt.show()

```
